```python
import pandas as pd
```

```python
ncd = pd.read_excel("data_ncd.xlsx")
overweight = pd.read_excel("data_overweight.xlsx")
underweight = pd.read_excel("data_underweight.xlsx")
water = pd.read_excel("data_water.xlsx")
pollution = pd.read_excel("data_pollution.xlsx")

raised_cholesterol = pd.read_excel("data_raised_cholesterol.xlsx")
fasting_blood = pd.read_excel("data_fasting_blood.xlsx")
raised_bp = pd.read_excel("data_raised_bp.xlsx")
phys_activity = pd.read_excel("data_phys_activity.xlsx")
population = pd.read_csv("population.csv")
```

```python
ncd
```

```python
it1 = ncd.merge(overweight)
it2 = it1.merge(underweight)
it3 = it2.merge(water)
it4 = it3.merge(population)
it5 = it4.merge(pollution)
it6 = it5.merge(raised_cholesterol)
it7 = it6.merge(fasting_blood)
it8 = it7.merge(raised_bp)
it9 = it8.merge(phys_activity)
data = it9
# it5.to_csv("full_data.csv", index=False)
```

```python
# data = pd.read_csv('full_data.csv')
data = data.dropna()
```

```python
data.loc[data['sex']=='Female', 'population'] = data.loc[data['sex']=='Female', 'population']/2
data.loc[data['sex']=='Male', 'population'] = data.loc[data['sex']=='Male', 'population']/2
```

```python
data.head()
```

```python
data['ncd_deaths'] = data['deaths']/data['population']
data['underweight_pc'] = data['underweight']/100*data['population']
data['overweight_pc'] = data['overweight']/100*data['population']
data['water_pc'] = data['water']/100*data['population']
data['pollution_pc'] = data['pollution']/100*data['population']
data['raised_cholesterol_pc'] = data['raised_cholesterol']/100*data['population']
data['fasting_blood_pc'] = data['fasting_blood']/100*data['population']
data['raised_bp_pc'] = data['raised_bp']/100*data['population']
data['phys_activity_pc'] = data['phys_activity']/100*data['population']


data['pollution'] = data['pollution_pc']*1000/data['population']
data['underweight'] = data['underweight_pc']*1000/data['population']
data['overweight'] = data['overweight_pc']*1000/data['population']
data['water'] = data['water_pc']*1000/data['population']
data['raised'] = data['raised_cholesterol_pc']*1000/data['population']
data['fasting'] = data['fasting_blood_pc']*1000/data['population']
data['raised_bp'] = data['raised_bp_pc']*1000/data['population']
data['phys_activity'] = data['phys_activity_pc']*1000/data['population']
```

```python
final_data = data[['country','sex','disease','ncd_deaths','underweight','overweight','water','pollution',
                   'raised_cholesterol','fasting_blood','raised_bp','phys_activity']]
```

```python
final_data.to_csv('full_data.csv', index=False)
```
